# **3.3-Making use of lists**

```python
# Define a list of integers
my_list = [10, 1, 8, 3, 5]

# Initialize a variable 'total' to hold the sum of the numbers in the list
total = 0

# Iterate over each index in the range of the list's length
for i in range(len(my_list)):
    # Add the element at the current index 'i' to the 'total'
    total += my_list[i]

# After completing the loop, print the total sum of the list elements
print(total)
```

# **Reverse()**




```python
my_list = [10, 1, 8, 3, 5]
my_list.reverse()
print(my_list)
```

When you call **my_list.reverse()**, this method reverses the elements of my_list in place. This means that the first element becomes the last, the second element becomes the second-to-last, and so on.
The reverse operation is performed directly on the list itself, meaning the original list is modified and no new list is created.

Will it still be acceptable with a list containing ** 100 elements** ? No, it won't.

Can you use the for loop to do the same thing automatically, irrespective of the list's length? **Yes, you can**.

```python
# Define a list of integers
my_list = [10, 1, 8, 3, 5]

# Get the length of the list
length = len(my_list)

# Loop through the first half of the list
for i in range(length // 2):
    # Swap the element at position 'i' with the element at the mirrored position from the end
    # The expression length - i - 1 in the code calculates the index of the element from the
   # end of the list that corresponds to the current index i from the start.
    # This is used for swapping elements to reverse part of the list.
    my_list[i], my_list[length - i - 1] = my_list[length - i - 1], my_list[i]

# Print the modified list
print(my_list)
```

# **5-The bubble sort**

# **5.1-Sort()**

How many passes do we need to sort the entire list? (ascending order)

```python
my_list = []
swapped = True
num = int(input("How many elements do you want to sort: "))

for i in range(num):
    val = int(input("Enter a list element: "))
    my_list.append(val)

while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            swapped = True
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print("\nSorted:")
print(my_list)
```

```python
#With python is magic
my_list = []
num = int(input("How many elements do you want to sort: "))

for i in range(num):
    val = int(input("Enter a list element: "))
    my_list.append(val)
my_list.sort()
print(my_list)
```

# **5.2-Reverse=True**

Now with Descending Order

```python
my_list = []
swapped = True
num = int(input("How many elements do you want to sort: "))

for i in range(num):
    val = int(input("Enter a list element: "))
    my_list.append(val)

while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] < my_list[i + 1]: #we just reverse the assignment operators
            swapped = True
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print("\nSorted:")
print(my_list)
```

# **6-Powerful slices**

A slice is an element of Python syntax that allows you to make a brand new copy of a list, or parts of a list.

It actually copies the list's contents, not the list's name.
image.png

```python
# Copying the entire list.
list_1 = [1,2,3,5,8]
list_2 = list_1[:]
print(list_2)
```

```python
# Copying some part of the list.
my_list = [10, 8, 6, 4, 2]
new_list = my_list[1:3]
print(new_list)
```

To repeat:

start is the index of the first element included in the slice;

end is the index of the first element **not included in the slice.**

```python
my_list = [10, 8, 6, 4, 2]
new_list = my_list[1:-1]
print(new_list)
# Therefore, this slice includes the elements from the second element to the second-to-last element of the list.
```

If the start specifies an element lying further than the one described by the end (from the list's beginning point of view), the slice will be empty:

```python
my_list = [10, 8, 6, 4, 2]
new_list = my_list[-1:1]
print(new_list)
```

```python
my_list = [10, 8, 6, 4, 2]
new_list = my_list[:3]
print(new_list)
```

```python
new_list = my_list[3:]
print(new_list)
```

```python
 #makes a copy of the whole list:
new_list = my_list[:]
print(new_list)
```
